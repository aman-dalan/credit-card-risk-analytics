```python
df['Net_income']=df['Salary']-df['ExDebt (Liability)']
df_new=df.drop(['Application id','first_name', 'last_name', 'email','gender','address','Salary','ExDebt (Liability)','INT_ID','Prev_ID','AGT_ID','Booking','Booking_Amt'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
df_new['empstaus']= le.fit_transform(df_new['empstaus']) 
df_new['ExCus (Customer in Past)']= le.fit_transform(df_new['ExCus (Customer in Past)']) 
df_new['tdecision']= le.fit_transform(df_new['tdecision'])
df_new['Source']= le.fit_transform(df_new['Source'])
```

```python
df_new = df_new[df_new.tdecision != 2]
df_new
```

```python
df_new.corr()
```

```python
x = df_new[[ 'Net_income','age','ExCus (Customer in Past)','empstaus','Source']]
y = df_new['tdecision']
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
#df_new['Booking_Amt']=sc_x.fit_transform(df_new['Booking_Amt'])
#df_new['Net_income']=sc_x.fit_transform(df_new['Net_income'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C':[0.00001,0.0001,0.001,0.01,0.1,1,10,100,1000,10000]}
grid = GridSearchCV(model,parameters, cv=3)
grid=grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
classifier=LogisticRegression(C=1000)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, y_pred)
```
